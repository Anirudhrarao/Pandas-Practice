### What is a Series?
* A pandas series is like a column in table.
* It is 1-d array

```python
a = [1,2,3,4,5,True]
```

```python
myvar = pd.Series(a)
myvar
```

### Label
* If nothing is specified then the value are labeled with thier index number.

```python
myvar[0]
```

### Create labels with index

```python
myvar = pd.Series(a,index=['First','Second','Thirs','Four','Five'])
myvar
```

```python
myvar['First']
```

### Key/Value Objects as Series
* You can also use key/value object or dict as series.

```python
calories = {'day1':420,'day2':320,'day3':330}
```

```python
myvar = pd.Series(calories)
```

```python
myvar
```

### DataFrames
* Data sets in pandas are usually 2-d or multi-dimensional tables, called as DataFrames.

```python
data = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}
```

```python
myvar = pd.DataFrame(data)
```

```python
myvar
```

```python
myvar = pd.DataFrame(data,index=['First_Day','Second_Day', 'Third_Day'])
```

```python
myvar
```

* Pandas use the "Loc" attribute to return one or more specified row(s).
